```python
import networkx as nx
import numpy as np
import matplotlib
```

```python
g = nx.Graph()
g.add_nodes_from(['A','B','C','D','E','F','G','H'])
g.add_edges_from([('A', 'B'),('B','C'),('B','D'),('C','E'),('D','G'),('D','E'),('E','F'),('H','G'),('E','G'),('E','H')])
nx.draw_networkx(g, node_color ='green')
```

```python
total_node=len(g.nodes)
given_edges=len(g.edges)
print("Nodes:",total_node," Edges:",given_edges)
```

```python
adj_list={}
for node in g.nodes():
    adj_list[node]=[]
    for nei in g.neighbors(node):
        adj_list[node].append(nei)
print(adj_list)
```

```python
print (nx.adjacency_matrix(g))
```

```python
total_node=len(g.nodes)
total_edges=total_node*(total_node-1)
print(len(g.edges)/total_edges)
```

```python
for node in g.nodes(): 
    print(node, g.degree(node))
```

```python
g_copy=g.copy()
g_copy.remove_node('A')
print(g_copy.nodes())
```

## Graphs for Q2 & Q3

```python
g1= nx.DiGraph()
g1.add_nodes_from(['Page A','Page B','Page C','Page D'])
g1.add_edges_from([('Page A', 'Page B'),('Page A','Page C'),('Page D','Page C'),('Page C','Page A'),('Page B','Page C')])
nx.draw_networkx(g1, node_color ='green')
```

```python
g2= nx.DiGraph()
g2.add_nodes_from(['Page A','Page B','Page C','Page D'])
g2.add_edges_from([('Page A', 'Page B'),('Page B','Page A'),('Page B','Page C'),('Page C','Page A'),('Page D','Page A'),('Page C','Page D')])
nx.draw_networkx(g2, node_color ='green')
```

```python
g3= nx.DiGraph()
g3.add_nodes_from(['Home','About','Product','Links','External Site A','External Site B','External Site C','External Site D'])
g3.add_edges_from([('Home', 'About'),('Home', 'Product'),('Home', 'Links'),#('About','Home'),('Product','Home'),('Links','Home'),
                  ('Links','External Site A'),('Links','External Site B'),('Links','External Site C'),('Links','External Site D')])
nx.draw_networkx(g3, node_color ='green')
```
