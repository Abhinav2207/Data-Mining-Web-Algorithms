```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("movie.csv")
arr=df.to_numpy()
arr
```

```python
rows,cols=arr.shape
distance=np.zeros((rows,rows))
for i in range (rows):
    for j in range(rows):
        for k in range(cols):
            distance[i][j]+=(arr[i][k]-arr[j][k])**2
        distance[i][j]=distance[i][j]**0.5

print(distance.reshape(20,20))
```
